```python
from __future__ import division
%pylab inline
from scikits.samplerate import resample
```

Let's first plot the raw data as collected by Bertrand.

```python
figsize(14,8)
x = np.loadtxt('WFMSCL02.CSV', delimiter=',')
plot(x[:,0],x[:,1])
```

In order to be able to listen to it, we need to reduce the sample rate to something acceptable by the sound card, say 48kHz.

```python
x_48k = resample(x[:,1], 48e-3, 'sinc_best')
x_48k /= x_48k.max()
plot(x_48k)
```

Now let's listen to it.

```python
from IPython.display import Audio
```

```python
Audio(x_48k, rate=48000)
```

And look at its spectrogram.

```python
_ = specgram(x[:,1], NFFT=256, Fs=1e6)
ylabel('Freq [Hz]')
xlabel('Time [s]')
plt.tight_layout()
```
